## Adding Noise to our synthetic Marmousi, Overthrust, Homogenous Model

```python
import numpy as np

# Load your overthrust model data into a NumPy array
marmousi_data = numpy_array

# Define the noise level range
min_noise_level = 0.1
max_noise_level = 0.5

# Generate random noise with the same shape as the overthrust data
noise_levels = np.random.uniform(min_noise_level, max_noise_level, size=marmousi_data.shape)
noise = np.random.normal(loc=0, scale=noise_levels, size=marmousi_data.shape)

# Add the noise to the overthrust data
noisy_data = marmousi_data + noise
```

```python
import numpy as np

# Load your overthrust model data into a NumPy array
overthrust_data = numpy_array1

# Define the noise level range
min_noise_level = 0.1
max_noise_level = 0.5

# Generate random noise with the same shape as the overthrust data
noise_levels = np.random.uniform(min_noise_level, max_noise_level, size=overthrust_data.shape)
noise = np.random.normal(loc=0, scale=noise_levels, size=overthrust_data.shape)

# Add the noise to the overthrust data
noisy_data1 = overthrust_data + noise

```

```python
import numpy as np

# Load your overthrust model data into a NumPy array
homogenous_data = numpy_array2

# Define the noise level range
min_noise_level = 0.1
max_noise_level = 0.5

# Generate random noise with the same shape as the overthrust data
noise_levels = np.random.uniform(min_noise_level, max_noise_level, size=homogenous_data.shape)
noise = np.random.normal(loc=0, scale=noise_levels, size=homogenous_data.shape)

# Add the noise to the overthrust data
noisy_data2 = homogenous_data + noise

```
